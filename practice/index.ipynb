```python
import torch
a=torch.arange(0,40,1).reshape(2,5,4)
print(a)
print(a.sum(axis=0))
print(a.sum(axis=1))
print(a.sum(axis=2))
print(a.sum(axis=[0,1])) #先对0轴求和，再对1轴求和，也可以先对1轴求和，再对0轴求和

```

```python
# 维度的个数不发生变化
print(a.sum(axis=1,keepdim=True))
a.sum(axis=1,keepdim=True).shape # torch.Size([2, 1, 4])
```
